```python
import sqlite3
import pandas as pd
```

```python
#to use the sqlite3 module, we must create a connection object that represents the database
con = sqlite3.connect("sakila.db")
```

```python
#Run a basic SQL query -> set is as a comment 
sql_query = '''SELECT * FROM customer'''

#Use pandas to pass sql query using connection form SQLite3 -> read the sql query with pd.read_sql
df = pd.read_sql(sql_query, con)
```

# Make a function that automatically takes the query and returns a dataframe

```python
def sql_to_df(sql_query):
    df = pd.read_sql(sql_query, con)
    return df
```

Selecting multiple columns

```python
query = '''SELECT first_name, last_name FROM customer;'''
sql_to_df(query).head()
```

# SQL DISTINCT Statement

* only the unique values

Select distinct country_ids from the city table

```python
sql_query = '''SELECT DISTINCT(country_id) FROM city'''
sql_to_df(sql_query).head()
```

# SQL WHERE

SELECT column_name
FROM table
WHERE column_name (math_operator) desired_value;

Select all customer info from the 1st store

```python
query = '''SELECT * FROM customer WHERE store_id = 1'''
sql_to_df(query).head()
```

Select all customer info for Mary

```python
query = '''SELECT * FROM customer WHERE first_name = "MARY"'''
sql_to_df(query)
```

Select all films from 2006 that are rated R

```python
query = '''SELECT * FROM film WHERE release_year = 2006 AND rating = "R"'''
sql_to_df(query).head()
```

Select all films from R or PG

```python
query = '''SELECT * FROM film WHERE rating = "R" or rating = "PG"'''
sql_to_df(query).head()
```

# Aggregate functions

Count the number of customers

```python
query = '''SELECT COUNT(customer_id) FROM customer'''
sql_to_df(query)
```

Average replacement_cost for rating PG

```python
query = '''SELECT AVG(replacement_cost) FROM film WHERE rating = "PG"'''
sql_to_df(query)
```

# SQL Wildcards
- % substitute for 0 or more caharacters
- _ substitute for a single character 
- [character list] sets and ranges of characters to match

Select any customers whose name starts with an M

```python
query = '''SELECT * FROM customer WHERE last_name LIKE "M%"'''
sql_to_df(query).head()
```

Select any customers whose last name ends with ing

```python
query = '''SELECT * FROM customer WHERE last_name LIKE "%ing"'''
sql_to_df(query)
```

# Character list wildcard

- in Mysql: WHERE value LIKE '[charlist]%'
- in SQLite3: WHERE value GLOB '[charlist]*'

Select any customers whose first name begins with A or B

```python
query = '''SELECT * FROM customer WHERE first_name GLOB "[AB]*"'''
sql_to_df(query).head()
```

# SQL ORDER BY

SELECT column name FROM table ORDER BY column_name ASC/DESC

Select all customers and order results by last name

```python
query = '''SELECT last_name FROM customer ORDER BY last_name'''
sql_to_df(query).head()
```

Select all customers and order results by last name, descending

```python
query = '''SELECT last_name FROM customer ORDER BY last_name DESC'''
sql_to_df(query).head()
```

# SQL GROUP BY

SELECT column_name, aggregate_function(column_name) FROM table WHERE column_name operator value GROUP BY column_name;

Count the number of customers per store

```python
query = '''SELECT store_id, COUNT(customer_id) FROM customer GROUP BY store_id'''
sql_to_df(query)
```
